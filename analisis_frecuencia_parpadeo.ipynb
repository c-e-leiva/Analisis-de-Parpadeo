# 📊 Análisis de la Frecuencia de Parpadeo en Diferentes Contextos

Este proyecto explora cómo el entorno afecta la frecuencia de parpadeo, comparando un contexto normal con uno frente a la computadora. El parpadeo es esencial para la salud ocular, y su frecuencia puede verse alterada al pasar tiempo frente a pantallas. 🔍

## 🎯 Objetivo del Análisis
El objetivo de este análisis es evaluar y comparar la frecuencia de parpadeo en dos contextos: un ambiente normal y uno frente a la computadora. Se busca entender cómo el tiempo frente a pantalla impacta en la salud visual, promoviendo la importancia del cuidado ocular. 🖥️👀

## 📂 Obtención de los Datasets

Los datasets utilizados en este análisis se generaron mediante un script en Python y están disponibles en mi repositorio de GitHub para facilitar la reproducibilidad. 🔗

### 🔢 Nombres de los Datasets y Estructura
1. **registros_parpadeo_capturado**:
   - **Descripción**: Contiene registros de parpadeo capturados mientras estaba frente a la computadora durante 60 minutos, registrando cada parpadeo con fecha y hora.
   - **Columna**: `Fecha y Hora del Parpadeo Capturado`.

2. **registros_parpadeo_normal**:
   - **Descripción**: Simula la frecuencia de parpadeo en un ambiente sin pantalla, con parpadeos generados aleatoriamente entre 15 y 20 por minuto.
   - **Columna**: `Fecha y Hora del Parpadeo Normal`.

Estos conjuntos de datos serán la base para analizar cómo el tiempo frente a pantallas afecta nuestra frecuencia de parpadeo y, en consecuencia, nuestra salud ocular. 👓

Se cargan y visualizan los datasets para obtener una primera impresión.

# 📊 Comparación del Total de Parpadeos en 60 Minutos

# 👁️ Distribución de Tipos de Parpadeos
Este análisis busca comparar la frecuencia de dos tipos de parpadeo: aquellos que ocurren en condiciones normales y aquellos capturados mientras una persona está frente a una pantalla. La comparación ayuda a entender cómo cambia el patrón de parpadeo en situaciones específicas (como el uso de dispositivos electrónicos) en contraste con una condición normal.

```python
# Contar el total de parpadeos en cada dataset
total_capturados = capturados.shape[0]
total_normales = normales.shape[0]

# Crear listas para los tipos de parpadeo y sus totales
tipos_de_parpadeo = ['Parpadeo Capturado', 'Parpadeo Normal']
totales = [total_capturados, total_normales]

# Calcular el total general
total_general = total_capturados + total_normales

# Calcular porcentajes
porcentaje_capturados = (total_capturados / total_general) * 100
porcentaje_normales = (total_normales / total_general) * 100

# Función personalizada para mostrar el porcentaje con color y tamaño
def formato_porcentaje(pct):
    return f'{pct:.1f}%'

# Crear el gráfico de torta
plt.figure(figsize=(6, 6))
plt.pie(totales, labels=tipos_de_parpadeo, autopct=lambda p: f'{p:.2f}%', startangle=40, colors=['orange', 'black'], textprops=dict(color='red', fontsize=16))

# Título del gráfico
plt.title('Distribución de Tipos de Parpadeo: Capturados vs. Normales', fontsize=16)

# Mostrar el gráfico
plt.axis('equal')  # Para que el gráfico sea un círculo
plt.tight_layout()
plt.show()

# Imprimir la cantidad total de parpadeos y sus porcentajes
print(f'Cantidad Total de Parpadeos Capturados: {total_capturados}')
print(f'Porcentaje de Parpadeos Capturados: {porcentaje_capturados:.2f}%')
print(f'\nCantidad Total de Parpadeos Normales: {total_normales}')
print(f'Porcentaje de Parpadeos Normales: {porcentaje_normales:.2f}%')
```

📉 Análisis de la Distribución de Parpadeos

La distribución muestra que solo un 16.16% de los parpadeos ocurrieron en condiciones de pantalla, mientras que un 83.84% ocurrieron en un contexto normal. Esto resalta cómo la exposición a pantallas reduce significativamente la frecuencia de parpadeo.

# ⏳ Comparación en Intervalos de 10 Minutos
Se comparan los parpadeos normales y capturados organizados en intervalos de 10 minutos. Esto permite un análisis detallado de cómo evoluciona la frecuencia de parpadeo con el tiempo.

```python
# Función para cargar y procesar los datos
def procesar_datos(ruta_csv, columna_fecha, intervalo):
    df = pd.read_csv(ruta_csv)
    df[columna_fecha] = pd.to_datetime(df[columna_fecha])
    inicio = df[columna_fecha].min()
    fin = df[columna_fecha].max() + pd.Timedelta(minutes=10)
    intervalos = pd.date_range(start=inicio, end=fin, freq=intervalo)
    df['Intervalo'] = pd.cut(df[columna_fecha], bins=intervalos, right=False)
    totales = df.groupby('Intervalo').size().reset_index(name='Total')
    return totales

# Cargar y procesar los datasets de parpadeos normales y capturados
df_resultados_normales = procesar_datos(
    r"E:\proyecto\analisis_parpadeo\dataset\registros_parpadeo_normal.csv",
    'Fecha y Hora del Parpadeo Normal',
    '10T'  # Cambiar a intervalos de 10 minutos
)

df_resultados_capturados = procesar_datos(
    r"E:\proyecto\analisis_parpadeo\dataset\registros_parpadeo_capturado.csv",
    'Fecha y Hora del Parpadeo Capturado',
    '10T'  # Cambiar a intervalos de 10 minutos
)

# Imprimir los DataFrames finales por separado
print("\nDataFrame con Totales de Parpadeos Normales:")
print(df_resultados_normales)
print("\nDataFrame con Totales de Parpadeos Capturados:")
print(df_resultados_capturados)

# Imprimir la cantidad total de parpadeos normales y capturados
total_parpadeos_normales = df_resultados_normales['Total'].sum()
total_parpadeos_capturados = df_resultados_capturados['Total'].sum()
print(f"\nCantidad Total de Parpadeos Normales: {total_parpadeos_normales}")
print(f"Cantidad Total de Parpadeos Capturados: {total_parpadeos_capturados}")
```

## 📊 Visualización en Gráfico de Barras
Gráfico de barras para comparar visualmente los parpadeos normales y capturados por cada intervalo de tiempo.

```python
#Graficar
plt.figure(figsize=(14, 5))

# Crear un índice para los intervalos
indices = range(len(df_resultados_normales))

# Graficar parpadeos normales
bar_width = 0.4
plt.bar(indices, df_resultados_normales['Total'], width=bar_width, label='Parpadeos Normales', color='black', align='center')

# Graficar parpadeos capturados, desplazando el eje X
plt.bar([i + bar_width for i in indices], df_resultados_capturados['Total'], width=bar_width, label='Parpadeos Capturados', color='orange', align='center')

# Configuración de etiquetas y título
plt.xlabel('Intervalos de Tiempo')
plt.ylabel('Cantidad de Parpadeos')
plt.title('Comparación de Parpadeos Normales y Capturados por Intervalos de 10 Minutos')

# Crear etiquetas de intervalo
interval_labels = [f'Intervalo {i+1} ({i*10} - {(i+1)*10} min)' for i in indices]
plt.xticks([i + bar_width / 2 for i in indices], interval_labels, rotation=45)

# Agregar etiquetas con totales arriba de cada barra
for i in indices:
    plt.text(i, df_resultados_normales['Total'][i] + 0.8, str(df_resultados_normales['Total'][i]), ha='center', color='black')
    plt.text(i + bar_width, df_resultados_capturados['Total'][i] + 2, str(df_resultados_capturados['Total'][i]), ha='center', color='black')

# Mover la leyenda a la parte inferior izquierda
plt.legend(loc='lower left', bbox_to_anchor=(0, 0))

# Agregar líneas de cuadrícula en el eje Y
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ajustar el espacio para que la leyenda no se superponga
plt.tight_layout(rect=[0, 0, 0.85, 1])

# Mostrar el gráfico
plt.show()
```

### Análisis de la Comparación de Parpadeos Normales y Capturados en Intervalos de 10 Minutos

1. **Parpadeos Normales**: A lo largo de los intervalos de 10 minutos analizados, los parpadeos normales se presentan con una frecuencia bastante constante, oscilando entre 171 y 188. Este comportamiento sugiere que, en un entorno libre de distracciones y en condiciones de relajación, las personas mantienen un patrón de parpadeo natural y predecible. Esta estabilidad en la frecuencia de parpadeos puede interpretarse como una respuesta fisiológica saludable, donde el ojo se humecta adecuadamente y se protege de la fatiga, lo cual es crucial para el bienestar ocular a largo plazo.

2. **Parpadeos Capturados**: En contraste, los parpadeos capturados muestran una tendencia a la baja, comenzando con un pico de 80 en el primer intervalo y disminuyendo notablemente a lo largo de los siguientes periodos. Esta reducción puede reflejar un fenómeno de aumento en la concentración y la atención hacia la tarea en pantalla. A medida que los individuos se involucran más en la actividad, su necesidad de parpadear puede disminuir, posiblemente debido a un enfoque sostenido en el contenido visual. Esta disminución podría también estar relacionada con el fenómeno de la "fatiga visual", que es común durante actividades prolongadas frente a pantallas, donde el esfuerzo constante por mantener la fijación visual puede llevar a una reducción en la frecuencia de parpadeos.

3. **Implicaciones**: La notable diferencia entre los patrones de parpadeos normales y capturados no solo resalta el impacto del ambiente de trabajo en la salud ocular, sino que también sugiere la necesidad de implementar prácticas que fomenten descansos regulares durante el uso prolongado de pantallas. Esta observación es fundamental, ya que la reducción de parpadeos puede contribuir a la sequedad ocular, incomodidad y otros síntomas relacionados con el uso excesivo de dispositivos electrónicos. Por lo tanto, es recomendable promover una mayor conciencia sobre la importancia de parpadear con frecuencia y realizar pausas visuales, para mantener la salud ocular y el bienestar general durante el uso de tecnología.


# Serie Temporal de Parpadeos en Intervalos de 5 Minutos

Este análisis compara la frecuencia de parpadeos normales y capturados frente a una pantalla en intervalos de 5 minutos para entender patrones temporales. Se procesa y agrupa la información en intervalos de 5 minutos, y luego se visualiza como una serie temporal para observar posibles variaciones en la frecuencia de parpadeos.

### Carga y Procesamiento de Datos
En este código se cargan los datos de los registros de parpadeos normales y capturados y los organiza en intervalos de 5 minutos. Se genera un DataFrame df_intervalos que contiene el total de parpadeos normales y capturados por intervalo, usando etiquetas para cada intervalo de tiempo.

```python
# Función para cargar y procesar los datos
def procesar_datos(ruta_csv, columna_fecha, intervalo):
    df = pd.read_csv(ruta_csv)
    df[columna_fecha] = pd.to_datetime(df[columna_fecha])
    inicio = df[columna_fecha].min()
    fin = df[columna_fecha].max() + pd.Timedelta(minutes=5)
    intervalos = pd.date_range(start=inicio, end=fin, freq=intervalo)
    df['Intervalo'] = pd.cut(df[columna_fecha], bins=intervalos, right=False)
    totales = df.groupby('Intervalo').size().reset_index(name='Total')
    return totales

# Cargar y procesar los datasets de parpadeos normales y capturados
df_resultados_normales = procesar_datos(
    r"E:\proyecto\analisis_parpadeo\dataset\registros_parpadeo_normal.csv",
    'Fecha y Hora del Parpadeo Normal',
    '5T'  # Intervalo de 5 minutos
)

df_resultados_capturados = procesar_datos(
    r"E:\proyecto\analisis_parpadeo\dataset\registros_parpadeo_capturado.csv",
    'Fecha y Hora del Parpadeo Capturado',
    '5T'  # Intervalo de 5 minutos
)

# Crear DataFrame final con intervalos y totales
df_intervalos = pd.DataFrame({'Intervalo': [f'Intervalo {i+1}' for i in range(12)]})
totales_normales = dict(zip(range(12), df_resultados_normales['Total'].values))
totales_capturados = dict(zip(range(12), df_resultados_capturados['Total'].values))
df_intervalos['Total de Parpadeos Normales'] = df_intervalos.index.map(totales_normales).fillna(0).astype(int)
df_intervalos['Total de Parpadeos Capturados'] = df_intervalos.index.map(totales_capturados).fillna(0).astype(int)

# Imprimir el DataFrame final con totales por intervalo
print("\nDataFrame Final con Totales de Parpadeos:")
print(df_intervalos)

# Calcular y mostrar la cantidad total de parpadeos normales y capturados
total_normales = df_intervalos['Total de Parpadeos Normales'].sum()
total_capturados = df_intervalos['Total de Parpadeos Capturados'].sum()

print("\nCantidad Total de Parpadeos Normales:", total_normales)
print("Cantidad Total de Parpadeos Capturados:", total_capturados)
```

### Visualización de la Serie Temporal
En este bloque, se genera un gráfico de líneas para comparar los parpadeos normales y capturados en cada intervalo de 5 minutos. Se añaden etiquetas en cada punto del gráfico para facilitar la interpretación de los valores de parpadeos en cada intervalo.

```python
# Crear etiquetas de intervalo con sus rangos
interval_labels = [f'Intervalo {i+1} ({i*5}-{(i+1)*5} min)' for i in range(12)]
df_intervalos['Intervalo'] = interval_labels

# Crear el gráfico de líneas
plt.figure(figsize=(13, 6))
plt.plot(df_intervalos['Intervalo'], df_intervalos['Total de Parpadeos Normales'], marker='o', label='Parpadeos Normales', color='black', linewidth=4)
plt.plot(df_intervalos['Intervalo'], df_intervalos['Total de Parpadeos Capturados'], marker='o', label='Parpadeos Capturados', color='orange', linewidth=4)

# Agregar etiquetas de totales en los puntos
for i in range(len(df_intervalos)):
    plt.text(i, df_intervalos['Total de Parpadeos Normales'][i] + 1.5, str(df_intervalos['Total de Parpadeos Normales'][i]), ha='center', color='black')
    plt.text(i, df_intervalos['Total de Parpadeos Capturados'][i] + 2, str(df_intervalos['Total de Parpadeos Capturados'][i]), ha='center', color='red')

# Personalización del gráfico
plt.title('Serie Temporal de Parpadeos Normales y Capturados')
plt.xlabel('Intervalos de Tiempo')
plt.ylabel('Total de Parpadeos')
plt.xticks(rotation=35)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

### Análisis de Serie Temporal de Parpadeos en Intervalos de 5 Minutos

1. **Parpadeos Normales**: La frecuencia de parpadeos en condiciones relajadas se mantiene constante, indicando un comportamiento natural y cómodo en entornos sin distracciones.

2. **Parpadeos Capturados**: Durante la interacción con la pantalla, los parpadeos registrados muestran una tendencia a la disminución con el tiempo. Este fenómeno puede interpretarse como un aumento en la concentración, donde el enfoque en la tarea reduce la necesidad de parpadear.

3. **Implicaciones**: La diferencia entre los parpadeos normales y los capturados resalta cómo la exposición a pantallas puede alterar los patrones de parpadeo, sugiriendo que una atención sostenida puede llevar a una menor frecuencia de parpadeos, lo que podría tener implicaciones en la salud ocular.
